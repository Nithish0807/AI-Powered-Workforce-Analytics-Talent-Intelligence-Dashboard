```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# App layout settings
st.set_page_config(
    page_title="AI Workflow Analytics & Talent Intelligence Dashboard",
    page_icon="📊",
    layout="wide"
)

# Custom Styling for polished look
st.markdown("""
    <style>
        .main { background-color: #f9fafb; }
        .metric-card {
            background-color: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 6px rgba(0,0,0,0.05);
            border-left: 5px solid #4F46E5;
        }
    </style>
""", unsafe_allow_html=True)

# Optimized Data Loader
@st.cache_data
def load_dashboard_data():
    macro = pd.read_csv("ai_workforce_displacement_global_2020_2026.csv")
    jobs = pd.read_csv("fact_job_postings.csv")
    skills = pd.read_csv("bridge_job_skills.csv")
    return macro, jobs, skills

try:
    df_macro, df_jobs, df_skills = load_dashboard_data()
except Exception as e:
    st.error(f"Data Connection Error: Please verify that all CSV datasets are uploaded to Colab. Detailed error: {e}")
    st.stop()

# Interactive Sidebar Control Panel
st.sidebar.title("🤖 Navigation Hub")
st.sidebar.markdown("Switch modules to track high-level workplace changes or granular talent demands.")
app_mode = st.sidebar.radio("Select Analytics Module:", ["🌐 AI Workflow Analytics", "🎯 Talent Intelligence Hub"])

# --- MODULE 1: AI WORKFLOW ANALYTICS (Macro View) ---
if app_mode == "🌐 AI Workflow Analytics":
    st.title("🌐 AI-Powered Workflow Analytics")
    st.markdown("### Strategic Overview: Global Automation Trends & Sector Workforce Displacement")
    st.write("---")

    sectors = sorted(df_macro['industry_sector'].dropna().unique())
    selected_sector = st.sidebar.selectbox("Filter Industry Sector Focus:", ["All Sectors"] + sectors)

    macro_filtered = df_macro.copy()
    if selected_sector != "All Sectors":
        macro_filtered = macro_filtered[macro_filtered['industry_sector'] == selected_sector]

    mcol1, mcol2, mcol3, mcol4 = st.columns(4)
    with mcol1:
        st.metric("Avg Automation Risk Score", f"{macro_filtered['sector_automation_risk_score'].mean():.2f}")
    with mcol2:
        st.metric("Avg Workforce Displaced", f"{macro_filtered['pct_sector_workforce_displaced'].mean() * 100:.2f}%")
    with mcol3:
        st.metric("New Roles Created", f"{macro_filtered['pct_sector_workforce_new_roles_created'].mean() * 100:.2f}%")
    with mcol4:
        net_change = macro_filtered['net_workforce_change_pct'].mean() * 100
        st.metric("Net Workforce Shift", f"{net_change:.2f}%", delta=f"{net_change:.2f}%")

    st.write("###")

    vcol1, vcol2 = st.columns(2)
    with vcol1:
        st.markdown("#### ⏳ Workforce Adjustment Evolution Over Time")
        timeline_df = macro_filtered.groupby(['quarter_label']).agg({
            'pct_sector_workforce_displaced': 'mean',
            'pct_sector_workforce_new_roles_created': 'mean'
        }).reset_index()
        fig_timeline = go.Figure()
        fig_timeline.add_trace(go.Scatter(x=timeline_df['quarter_label'], y=timeline_df['pct_sector_workforce_displaced']*100, mode='lines+markers', name='Displaced %', line=dict(color='#EF4444', width=3)))
        fig_timeline.add_trace(go.Scatter(x=timeline_df['quarter_label'], y=timeline_df['pct_sector_workforce_new_roles_created']*100, mode='lines+markers', name='New Roles Created %', line=dict(color='#10B981', width=3)))
        fig_timeline.update_layout(template="plotly_white", margin=dict(l=10, r=10, t=10, b=10))
        st.plotly_chart(fig_timeline, use_container_width=True)

    with vcol2:
        st.markdown("#### ⚡ Industry Automation Risk vs. AI Tool Adoption")
        fig_scatter = px.scatter(macro_filtered, x="sector_automation_risk_score", y="ai_tool_adoption_pct", color="income_group", size="ai_cited_layoff_announcements", hover_name="country", color_discrete_sequence=px.colors.qualitative.Safe)
        fig_scatter.update_layout(template="plotly_white", margin=dict(l=10, r=10, t=10, b=10))
        st.plotly_chart(fig_scatter, use_container_width=True)

# --- MODULE 2: TALENT INTELLIGENCE HUB (Micro View) ---
else:
    st.title("🎯 Talent Intelligence & Demand Hub")
    st.markdown("### Operational Insights: Skills Demand Mapping & Competitive Benchmarking")
    st.write("---")

    exp_levels = sorted(df_jobs['experience_level'].dropna().unique())
    selected_exp = st.sidebar.multiselect("Filter Experience Levels:", exp_levels, default=exp_levels)
    jobs_filtered = df_jobs[df_jobs['experience_level'].isin(selected_exp)]

    jcol1, jcol2, jcol3 = st.columns(3)
    with jcol1:
        st.metric("Total Tracked Open Postings", f"{len(jobs_filtered):,}")
    with jcol2:
        st.metric("Average Maximum Market Salary", f"${jobs_filtered['salary_max_usd'].mean():,.0f} USD")
    with jcol3:
        st.metric("Total Cumulative Job Applicants", f"{jobs_filtered['applicants_estimate'].sum():,}")

    st.write("###")

    vcol3, vcol4 = st.columns(2)
    with vcol3:
        st.markdown("#### 🔥 Top 10 Most Demanded Tech Skills")
        merged_skills = pd.merge(jobs_filtered, df_skills, on="job_id")
        top_skills = merged_skills['skill'].value_counts().reset_index().head(10)
        top_skills.columns = ['Skill', 'Postings Demand']
        fig_skills = px.bar(top_skills, x="Postings Demand", y="Skill", orientation="h", color="Postings Demand", color_continuous_scale="blues")
        fig_skills.update_layout(template="plotly_white", yaxis={'categoryorder':'total ascending'}, margin=dict(l=10, r=10, t=10, b=10))
        st.plotly_chart(fig_skills, use_container_width=True)

    with vcol4:
        st.markdown("#### 🏢 Top Enterprise Hiring Drivers")
        top_companies = jobs_filtered['company'].value_counts().reset_index().head(10)
        top_companies.columns = ['Company', 'Job Openings']
        fig_company = px.bar(top_companies, x="Company", y="Job Openings", color="Job Openings", color_continuous_scale="purples")
        fig_company.update_layout(template="plotly_white", margin=dict(l=10, r=10, t=10, b=10))
        st.plotly_chart(fig_company, use_container_width=True)

    st.write("---")
    st.markdown("#### 📑 Active Searchable Talent Pool Matrix")
    display_cols = ['job_title', 'company', 'country', 'experience_level', 'remote_option', 'salary_max_usd']
    st.dataframe(jobs_filtered[display_cols].head(100), use_container_width=True)
```
