```python
import sqlite3
import pandas as pd


conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print("Connected to in-memory SQLite database.")
```

```python
create_table_sql = """
CREATE TABLE Employees (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    department TEXT NOT NULL,
    salary REAL NOT NULL,
    hire_date TEXT NOT NULL,
    is_active INTEGER
);
"""
cursor.execute(create_table_sql)
conn.commit()
print("Table 'Employees' created successfully.")
```

```python
def fetch_and_display(query, description):
    print(f"\n--- {description} ---")
    cursor.execute(query)
    rows = cursor.fetchall()
    columns = [description[0] for description in cursor.description]
    df = pd.DataFrame(rows, columns=columns)
    display(df)


fetch_and_display("SELECT * FROM Employees;", "All Employees")


fetch_and_display("SELECT name, department, salary FROM Employees;", "Name, Department, Salary")


fetch_and_display("SELECT * FROM Employees WHERE department = 'IT';", "Employees in IT Department")


fetch_and_display("SELECT name, salary FROM Employees ORDER BY salary DESC;", "Employees by Salary (Descending)")


fetch_and_display("SELECT name, salary FROM Employees ORDER BY salary DESC LIMIT 5;", "Top 5 Highest Paid Employees")


fetch_and_display("SELECT DISTINCT department FROM Employees;", "Unique Departments")
```

```python
fetch_and_display("SELECT name, department FROM Employees WHERE name LIKE 'A%';", "Employees with names starting with 'A'")

fetch_and_display("SELECT name, salary FROM Employees WHERE salary BETWEEN 60000 AND 70000 ORDER BY salary;", "Employees with Salary between 60k and 70k")


fetch_and_display("SELECT name, department FROM Employees WHERE department IN ('Sales', 'HR');", "Employees in Sales or HR")


fetch_and_display("SELECT name, department, salary FROM Employees WHERE is_active = 1 AND department = 'IT' AND salary > 80000;", "Active IT Employees with Salary > 80k")


fetch_and_display("SELECT name, department FROM Employees WHERE NOT department = 'Marketing';", "Employees NOT in Marketing")
```

```python
fetch_and_display("SELECT COUNT(*) AS total_employees FROM Employees;", "Total Number of Employees")

fetch_and_display("SELECT SUM(salary) AS total_salary FROM Employees;", "Total Salary Bill")


fetch_and_display("SELECT department, AVG(salary) AS average_salary FROM Employees GROUP BY department;", "Average Salary by Department")


fetch_and_display("SELECT department, MAX(salary) AS max_salary, MIN(salary) AS min_salary FROM Employees GROUP BY department;", "Max and Min Salary by Department")


fetch_and_display("SELECT department, AVG(salary) AS average_salary FROM Employees GROUP BY department HAVING AVG(salary) > 70000;", "Departments with Average Salary > 70k")
```

```python
case_query = """
SELECT
    name,
    salary,
    CASE
        WHEN salary >= 80000 THEN 'High Salary'
        WHEN salary >= 60000 THEN 'Medium Salary'
        ELSE 'Low Salary'
    END AS salary_category
FROM Employees
ORDER BY salary DESC;
"""
fetch_and_display(case_query, "Employees with Salary Categories using CASE")
```

```python

subquery_example = """
SELECT name, salary
FROM Employees
WHERE salary > (SELECT AVG(salary) FROM Employees);
"""
fetch_and_display(subquery_example, "Employees Earning More Than Average Salary (Subquery)")


create_view_sql = """
CREATE VIEW ActiveEmployees AS
SELECT id, name, department, salary, hire_date
FROM Employees
WHERE is_active = 1;
"""
cursor.execute(create_view_sql)
conn.commit()
print("View 'ActiveEmployees' created successfully.")


fetch_and_display("SELECT * FROM ActiveEmployees;", "Querying the ActiveEmployees View")
```

```python
create_index_sql = "CREATE INDEX idx_department ON Employees (department);"
cursor.execute(create_index_sql)
conn.commit()
print("Index 'idx_department' created on 'Employees' table.")

alter_table_add_col_sql = "ALTER TABLE Employees ADD COLUMN email TEXT;"
cursor.execute(alter_table_add_col_sql)
conn.commit()
print("Column 'email' added to 'Employees' table.")

update_email_sql = "UPDATE Employees SET email = LOWER(REPLACE(name, ' ', '.')) || '@example.com' WHERE id IN (1, 2, 3);"
cursor.execute(update_email_sql)
conn.commit()
print("Emails updated for some employees.")

fetch_and_display("SELECT id, name, email FROM Employees WHERE id IN (1, 2, 3);", "Employees with updated email")
```

```python
update_sql = "UPDATE Employees SET salary = salary * 1.10 WHERE department = 'IT';"
cursor.execute(update_sql)
conn.commit()
print("Salaries updated for IT department employees.")

fetch_and_display("SELECT name, department, salary FROM Employees WHERE department = 'IT';", "IT Employees after Salary Update")

delete_sql = "DELETE FROM Employees WHERE is_active = 0;"
cursor.execute(delete_sql)
conn.commit()
print("Inactive employees deleted.")

fetch_and_display("SELECT name, is_active FROM Employees WHERE is_active = 0;", "Remaining Inactive Employees (should be empty)")
fetch_and_display("SELECT COUNT(*) FROM Employees;", "Total Employees After Deletion")
```

```python
drop_table_sql = "DROP TABLE IF EXISTS Employees;"
cursor.execute(drop_table_sql)
conn.commit()
print("Table 'Employees' dropped successfully.")


drop_view_sql = "DROP VIEW IF EXISTS ActiveEmployees;"
cursor.execute(drop_view_sql)
conn.commit()
print("View 'ActiveEmployees' dropped successfully.")

conn.close()
print("Database connection closed.")
```
