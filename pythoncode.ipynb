```python


class Student:

    college_name = "ABC Engineering College"

    def __init__(self, student_id, name, department):
        self.student_id = student_id
        self.name = name
        self.department = department
        self.subjects = []
        self.marks = {}

    def add_subject(self, subject, mark):
        self.subjects.append(subject)
        self.marks[subject] = mark

    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        return sum(self.marks.values()) / len(self.marks)

    def grade(self):
        avg = self.calculate_average()

        if avg >= 90:
            return "A+"
        elif avg >= 80:
            return "A"
        elif avg >= 70:
            return "B"
        elif avg >= 60:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("=" * 50)
        print("College :", Student.college_name)
        print("Student ID :", self.student_id)
        print("Name :", self.name)
        print("Department :", self.department)

        print("\nSubjects and Marks")

        for subject in self.subjects:
            print(f"{subject} : {self.marks[subject]}")

        print("Average :", round(self.calculate_average(), 2))
        print("Grade :", self.grade())
        print("=" * 50)

#tuple

departments = (
    "CSE",
    "IT",
    "ECE",
    "EEE",
    "MECH"
)

print("Departments Available")

for dept in departments:
    print(dept)

print()



# Creating Objects


student1 = Student(101, "Nithish", "CSE")
student2 = Student(102, "Rahul", "IT")
student3 = Student(103, "Priya", "ECE")



# Adding Subjects


student1.add_subject("Python", 95)
student1.add_subject("SQL", 88)
student1.add_subject("AI", 92)

student2.add_subject("Python", 80)
student2.add_subject("SQL", 75)
student2.add_subject("AI", 70)

student3.add_subject("Python", 99)
student3.add_subject("SQL", 97)
student3.add_subject("AI", 98)


# List of Objects


students = [
    student1,
    student2,
    student3
]


# Display Student Details


print("\nSTUDENT DETAILS\n")

for student in students:
    student.display()


# Dictionary Example


department_count = {}

for student in students:

    dept = student.department

    if dept in department_count:
        department_count[dept] += 1
    else:
        department_count[dept] = 1


print("\nDepartment Wise Student Count")

for dept, count in department_count.items():
    print(dept, ":", count)



# Highest Average Student


highest = students[0]

for student in students:
    if student.calculate_average() > highest.calculate_average():
        highest = student

print("\nTop Performer")
print("Name :", highest.name)
print("Average :", round(highest.calculate_average(), 2))


# Search Student


search_id = 102

print("\nSearching Student ID :", search_id)

found = False

for student in students:

    if student.student_id == search_id:
        student.display()
        found = True
        break

if not found:
    print("Student Not Found")



# Total Students


print("\nTotal Students :", len(students))


# Average Marks of Entire Class


total = 0

for student in students:
    total += student.calculate_average()

class_average = total / len(students)

print("Class Average :", round(class_average, 2))



# Students Passed


print("\nStudents Passed")

for student in students:
    if student.grade() != "Fail":
        print(student.name)



# Students Scoring Above 90


print("\nStudents Above 90 Average")

for student in students:
    if student.calculate_average() >= 90:
        print(student.name)



# Ranking Students


sorted_students = sorted(
    students,
    key=lambda x: x.calculate_average(),
    reverse=True
)

print("\nStudent Ranking")

rank = 1

for student in sorted_students:
    print(rank, "-", student.name, "-", round(student.calculate_average(), 2))
    rank += 1
```
