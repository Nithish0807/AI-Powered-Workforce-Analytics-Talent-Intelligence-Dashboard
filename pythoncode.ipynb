```python
import pandas as pd

# Step 1: Create Dataset
mobiles = {
    "Mobile_ID": [101,102,103,104,105,106,107,108],
    "Brand": ["Samsung","Apple","OnePlus","Xiaomi","Vivo","Oppo","Realme","Motorola"],
    "Model": ["S24","iPhone 16","12R","Redmi Note 14","V50","Reno 13","GT 7","Edge 60"],
    "RAM_GB": [8,8,12,6,8,12,12,8],
    "Storage_GB": [128,256,256,128,256,256,512,128],
    "Price": [74999,89999,42999,18999,32999,38999,44999,27999],
    "Stock": [15,10,20,30,25,18,12,22]
}

df = pd.DataFrame(mobiles)

print("========== MOBILE STORE DATA ==========")
print(df)

print("\nTotal Number of Mobiles:")
print(len(df))

print("\nMost Expensive Mobile:")
print(df.loc[df["Price"].idxmax()])

print("\nCheapest Mobile:")
print(df.loc[df["Price"].idxmin()])

print("\nAverage Mobile Price:")
print(df["Price"].mean())

print("\nMobiles Costing More Than ₹40,000:")
print(df[df["Price"] > 40000])

print("\nMobiles with 12 GB RAM:")
print(df[df["RAM_GB"] == 12])

df["Inventory_Value"] = df["Price"] * df["Stock"]

df["Discount"] = df["Price"] * 0.10

df["Final_Price"] = df["Price"] - df["Discount"]

print("\nDataset After Adding New Columns:")
print(df)

print("\nMobiles Sorted by Price:")
print(df.sort_values(by="Price", ascending=False))

print("\nTotal Stock Available:")
print(df["Stock"].sum())

highest = df.loc[df["Price"].idxmax()]
print("\nHighest Priced Brand:")
print(highest["Brand"], "-", highest["Model"], "- ₹", highest["Price"])
```
